### 1.0 Introduction
The Kenyan government has carried out a population census every 10 years since 1948. The latest census was done in 2019 and it recorded a population of 47.5 million. This project utilizes the population and housing census data of 2019 to explore the Kenyan population characteristis.This is achieved by plotting visualizations with the datasets. The data is composed of several sets and this project utilizes just a few because of time limitation. The datasets were obtained from the africaopendata.org portal. Four datasets were used namely: 'Kenya population density per county per Land area','Distribution of population by Sex and County','Distribution of population, number of Housholds and Average household size by county','Distribution of Population by Age and Sex'

### 3.0 Data Visualization
In this section, i plot graphs with the clean datasets. I also note some of the insights that become visible from the graphs.

```python
# importing library for plotting graphs
import matplotlib.pyplot as plt
```

###### i. Population Size per county

```python
bins=[500000,1000000,1500000,2000000,2500000,3000000,3500000,4000000,4500000]
df.plot(kind='bar', x='County_Name',y='Pop_Size', width=0.5)
#plt.bar(df['County_Name'],df['Pop_Size'],width=0.5)
plt.yticks(bins)
plt.title ('Population Size per County')
plt.xlabel('Counties')
plt.ylabel('Population Size (million)')
plt.figure(figsize=(300,6))
#plt.xticks(fontsize=0.5)
#plt.xticks(rotation=25)


#plt.tick_params(axis='x', which='major', labelsize=3)
plt.show()
```

from the bar graph the most populous county in Kenya is Nairobi, followed by Kiambu, Nakuru and Kakamega
The least populous county is Lamu, followed by isiolo, Tana River,Taita-Taveta,samburu, Tharaka Nithi.
It is interesting to note that Tharaka Nithi county is among the least populous counties contrary to popular belief

###### ii. Comparison of county sizes

```python
#Land area per county
df.plot(kind='bar', x='County_Name',y='Land_Area(Sq.Km)', width=0.5)
plt.title('Size of Counties (Land Area in Sqkm)')
plt.xlabel('County')
plt.ylabel('Pop. Density (Sqkm)')
plt.show()

```

Turkana and Marsabit counties cover the largest area in sqm followed by wajir, Garissa and Tana river.
Mombasa,Nairobi and Vihiga counties covers the least land area in sqkm even though from the earlier graph  Nairobi 
is the most populous county 

###### iii. Comparison of Pop. Density per county

```python
#Population (density per sqkm) per county
df.plot(kind='bar', x='County_Name',y='Pop_density_per_Sqkm', width=0.5)
plt.title('Population Density per County per sqkm')
plt.xlabel('County')
plt.ylabel('Pop. Density (Sqkm)')
plt.show ()
```

Mombasa and Nairobi are the most densely populated counties; This was to be expected 
since from my earlier Vizualizations they are the most populous but cover the least land area 
per sqkm
Tana river and Marsabit have the lowest population density (disappearing from view) followed by Lamu, Taita taveta, Garisss,wajir,
Mandera,isiolo, Turkana, West pokot, Samburu, Baringo, Laikipia, Narok,Kajiado;
On Revisting the data,Tana river and Marsabit have less than 10 people per sq km

###### iv. No. of Households per county

```python
df.plot(kind='bar', x='County_Name',y='No_of_Households', width=0.5)
plt.title('No. Of Households per county')
plt.xlabel('County')
plt.ylabel('No. of Households')
plt.show ()
```

Nairobi has the most no. of households followed by kiambu,Nakuru and Meru counties while the lowest no. of households
are in lamu, isiolo,samburu
The graph is an exact replica of the pop. size per county graph; hence suggesting a linear r/ship between pop. size and 
number of households; let's check that with the next graph

```python
df.plot(kind='line', x='Pop_Size',y='No_of_Households')
plt.title('Pop. Size vs No. Of Households')
plt.xlabel('Pop. Size')
plt.ylabel('No. of Households')
plt.show ()
```

As expected, graph shows that no. of household increases with pop. size

###### v. Average Household Size per county

```python
# Avg. Household size
df.plot(kind='bar', x='County_Name',y='Avg_Household_Size', width=0.5)
plt.title('Avg Household Size')
plt.xlabel('County')
plt.ylabel('Avg Household Size')
plt.show ()
```

The graph shows that across many counties,there is not a huge disparity in the average household size.
However some of the counties that had very low pop. sizes seem to have the 
largest household sizes; mandera,wajir,west pokot.
Nairobi County (which is the most densely populated and has the highest pop.size) has the least avg. household size

###### vii. National gender proportions

```python
ageandsex=pd.read_csv('ageandsex.csv')
```

```python
ageandsex
```

```python
ageandsex.dtypes
```

```python
ageandsex
```

```python
ageandsex.dtypes
```

```python
#extracting the total values for male nd female from above dataframe and creating a series for plotting a pie chart
male=ageandsex['Male'].iloc[0]
female=ageandsex['Female'].iloc[0]
male
lst=[male,female]
gender=pd.DataFrame(lst)

gender.insert(loc=0,column='Sex',value=('Male','Female'))
gender[0]=(gender[0]/(male+female))*100
gender
```

```python
gender[0].plot(kind='pie',labels=gender['Sex'],autopct='%1.2f%%',figsize=(8,8),fontsize=10, subplots='True')
plt.title('National gender proportions')
```

The national female-male proportions are almost equal 
contrary to a common notion that females are significantly more than males
